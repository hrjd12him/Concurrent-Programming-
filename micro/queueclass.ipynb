### using Multiprocessing Queue for the shared memory

* Shared memory allows multiple processes to access the same chunk of memory directly, while queues provide a safe way to exchange data between processes.

```python
import multiprocessing

def square(arr,q):
    for i in arr:
        q.put(i**2)

if __name__ == "__main__":
    arr = [10,20,30]
    q = multiprocessing.Queue()

    p1 = multiprocessing.Process(target=square,args=(arr,q))
    p1.start()
    p1.join()

    while q.empty() is False:
        print(q.get())
```
